```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb
```

```python
import os
print(os.listdir("Documents/Project"))
```

```python
df = pd.read_csv("Documents/Project/nigeria_houses_data.csv")
df.head()
print(df.shape)
df.info()
```

```python
print(df["title"].value_counts())
print(df["state"].value_counts())
```

```python
def format_naira(price):
    return (f"#{price:,.0f}")
```

```python
df. groupby("state")["price"].mean().sort_values(ascending=False).apply(format_naira)
```

```python
df.groupby("title")["price"].mean().sort_values(ascending=False).apply(format_naira)
```

```python
print(f"The average price: {format_naira(df["price"].mean())}")
print(f"The most expensive:{format_naira(df["price"].max())} ")
print(f"The cheapest:{format_naira(df["price"].min())} ")
print(f"The median:{format_naira(df["price"].median())} ")
```

```python
df.sort_values("price", ascending=False)
```

```python
print(df["price"].quantile([0.01,0.25,0.50,0.75,0.99]))
```

```python
df_clean= df[(df["price"]>=df["price"].quantile(0.01)) & (df["price"]<=df["price"].quantile(0.99))]
print(f"Length before cleaning: {len(df)}")
print(f"Length after cleaning: {len(df_clean)}")
```

```python
print(f"Average price:     {format_naira(df_clean['price'].mean())}")
print(f"Median price:      {format_naira(df_clean['price'].median())}")
print(f"Cheapest:          {format_naira(df_clean['price'].min())}")
print(f"Most expensive:    {format_naira(df_clean['price'].max())}")
```
